```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_llama = pd.read_csv('Llama-3-8B-Instruct-zero-shot-24-permutations - Sheet1.csv')
df_mistral = pd.read_csv('./Mistral-7B-Instruct-v0.3-24-permutations - Sheet1 (1).csv')
df_gemma = pd.read_csv('gemma-2-9b-it-24-permutations - Sheet1.csv')

new_column = [i % 4 for i in range(len(df_llama))]
df_llama['position'] = new_column
df_gemma['position'] = new_column
df_mistral['position'] = new_column
```

```python

# Define the binary list
binary_list = [0.91, 0.46, 0.85]

# Filter for permutations that are 'TRUE'
df_llama = df_llama[df_llama['Permutation'] == 'TRUE']
df_mistral = df_mistral[df_mistral['Permutation'] == 'TRUE']
df_gemma = df_gemma[df_gemma['Permutation'] == 'TRUE']

# Group by 'position' and calculate mean accuracy for each dataframe
grouped_llama = df_llama.groupby('position')['Accuracy'].mean().reset_index()
grouped_mistral = df_mistral.groupby('position')['Accuracy'].mean().reset_index()
grouped_gemma = df_gemma.groupby('position')['Accuracy'].mean().reset_index()

# Append the corresponding values from the binary list
grouped_llama = pd.concat([grouped_llama, pd.DataFrame({'position': ['Binary List Value'], 'Accuracy': [binary_list[0]]})], ignore_index=True)
grouped_mistral = pd.concat([grouped_mistral, pd.DataFrame({'position': ['Binary List Value'], 'Accuracy': [binary_list[1]]})], ignore_index=True)
grouped_gemma = pd.concat([grouped_gemma, pd.DataFrame({'position': ['Binary List Value'], 'Accuracy': [binary_list[2]]})], ignore_index=True)

# Align positions (0, 1, 2, 3, and binary list value)
positions = ['0', '1', '2', '3', 'Binary List Value']

# Prepare data for plotting
accuracy_data = {
    'Llama': grouped_llama['Accuracy'].tolist(),
    'Mistral': grouped_mistral['Accuracy'].tolist(),
    'Gemma': grouped_gemma['Accuracy'].tolist()
}

# Create the plot
bar_width = 0.2
x = np.arange(len(positions))  # The label locations

plt.figure(figsize=(10, 6))

# Plot each group
plt.bar(x - bar_width, accuracy_data['Llama'], width=bar_width, label='Llama')
plt.bar(x, accuracy_data['Mistral'], width=bar_width, label='Mistral')
plt.bar(x + bar_width, accuracy_data['Gemma'], width=bar_width, label='Gemma')

# Customize the plot
plt.xlabel('Position and Binary List Value')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison Across Models by Position and Binary List Value')
plt.xticks(x, positions)
plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define the binary list
binary_list = [0.91, 0.46, 0.85]

df_llama = df_llama[df_llama['Permutation'] == 'TRUE']
df_mistral = df_mistral[df_mistral['Permutation'] == 'TRUE']
df_gemma = df_gemma[df_gemma['Permutation'] == 'TRUE']

# Group by 'position' and calculate mean accuracy for each dataframe
grouped_llama = df_llama.groupby('position')['Accuracy'].mean().reset_index()
grouped_mistral = df_mistral.groupby('position')['Accuracy'].mean().reset_index()
grouped_gemma = df_gemma.groupby('position')['Accuracy'].mean().reset_index()

# Example data
df_llama = pd.DataFrame({'position': ['Binary Classification','0', '1', '2', '3'], 'Accuracy': [binary_list[0],0.9, 0.7, 0.8, 0.85,]})
df_gemma = pd.DataFrame({'position': ['Binary Classification','0', '1', '2', '3'], 'Accuracy': [binary_list[1],0.7, 0.6, 0.75, 0.8]})
df_mistral = pd.DataFrame({'position': ['Binary Classification','0', '1', '2', '3'], 'Accuracy': [binary_list[2],0.6, 0.65, 0.7, 0.75]})

# Prepare data for plotting
accuracy_data = {
    'Llama': df_llama['Accuracy'].tolist(),
    'Gemma': df_gemma['Accuracy'].tolist(),
    'Mistral': df_mistral['Accuracy'].tolist(),
}

models = list(accuracy_data.keys())
num_models = len(models)
positions = df_llama['position'].tolist()
num_positions = len(positions)
print(num_positions)

# Create a single plot for all models
plt.figure(figsize=(12, 8))
base_color = '#c51926'
colors = ['#c51926', '#4e79a7', '#f28e2b', '#76b7b2', '#e15759']
# Bar width and spacing
bar_width = 0.8 / num_positions  # Adjust width to fit multiple bars in each model group
x = np.arange(num_models)  # Base positions for the models

# Plot each position's accuracies for each model
for i, position in enumerate(positions):
    # Offset x positions for each position's bars
    x_positions = x + i * bar_width - (bar_width * (num_positions - 1) / 2)  # Center bars for each position
    accuracies = [accuracy_data[model][i] for model in models]
    plt.bar(x_positions, accuracies, width=bar_width, color=colors[i],label=f'{"Position " if position != "Binary Classification" else ""}{position}')

# Customize the plot
# plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('True Label - Binary vs Zero Shot')
plt.xticks(x, models)  # Set x-axis ticks to model names
plt.legend(title='Position')
plt.tight_layout()  # Adjust layout to fit labels

# Show the combined plot
plt.show()
```
